## Initial Data Analysis

### Quality of the data ###

As a first exploration of the data let's import the train set and see their structure.

```python
import pandas as pd
import numpy as np

df = pd.read_csv('Train.csv')
df.head(10)
```

A first contact with the dataset shows us a total of 6 categorical variables (Item_Fat_Content, Item_Type, Outlet_Size, Outlet_Location_Type and Outlet_Type), 3 quantitative variables (Item_Weight, Item_MRP and Outlet_Establishment_Year) and 2 identifiers (Item_Identifier and Outlet_Identifier).

We can also see the presence of missing values in the weights of the product and the size of the establishment.

The identifiers will not be taken into account in the model and should therefore be removed from the DataFrame, but will be kept as separate variables for possible analysis later.

```python
#Extract identifiers
items_id = df['Item_Identifier'].copy()
outlets_id = df['Outlet_Identifier'].copy()

#Drop identifiers columns
df.drop('Item_Identifier', axis='columns', inplace=True)
df.drop('Outlet_Identifier', axis='columns', inplace=True)

#Show changes
df.head(5)
```

The identifiers are useful to understand the behavior of the products but especially of the stores, with them we can obtain the store that sells the most grouped by Tier or <a href="https://public.tableau.com/views/BigMartSales/SalesbyTypeofItem?:embed=y&:display_count=yes&publish=yes">Sales of each type of product in each outlet</a>.

![Sales by Type of Item](https://i.imgur.com/1qYXyNQ.jpg "Sales by Type of Item")

The existence of very similar sales between outlets (with certain exceptions) seems to indicate that these may not be related to outlet tier. This hypothesis will be corroborated by a future correlation analysis between the variables.

Continuing with the analysis, let's analyze the missing values in the features.

### Missing Data Analysis

Let's visualize the missing values.

```python
def get_missing_data_table(dataframe):
    total = dataframe.isnull().sum()
    percentage = dataframe.isnull().sum() / dataframe.isnull().count()
    
    missing_data = pd.concat([total, percentage], axis='columns', keys=['TOTAL','PERCENTAGE'])
    return missing_data.sort_values(by='TOTAL', ascending=False)


get_missing_data_table(df)
```

The missing values represent a large percentage of both variables, so it is important to find a way to fulfill them that makes sense with the intended objective.

Let's find out how many weightless products we have, and which outlets are undersized.

```python
#Working with a copy of the main DataFrame

#Concat id and weight
df_copy = pd.concat([items_id, df['Item_Weight']], axis='columns')
#Get null values
df_copy = df_copy[df['Item_Weight'].isnull()]
#Get how many nulls ids are
len(df_copy['Item_Identifier'].unique())
```

```python
#Working with a copy of the main DataFrame

#Concat id and size
df_copy = pd.concat([outlets_id, df['Outlet_Size']], axis='columns')
#Get null values
df_copy = df_copy[df['Outlet_Size'].isnull()]
#Get nulls ids
df_copy['Outlet_Identifier'].unique()
```

As a first alternative to solve these missing values we will use their identifiers to look for other records in the dataset where the value exists.

```python
#----Item Weight
#Concat id and weight
df_copy = pd.concat([items_id, df['Item_Weight']], axis='columns')
#Get null values
df_copy = df_copy[df['Item_Weight'].isnull()]
#Get how many nulls ids are
nulls = df_copy['Item_Identifier'].unique()

from scipy.stats import mode
for id in nulls:
    #Get value
    df_weight = df.loc[items_id == id, 'Item_Weight']
    value = mode(df_weight.values, nan_policy='omit').mode[0]
    
    if value == 0:
        value = np.nan #No results if mode = 0
    
    df.loc[(df_weight == np.nan).index, 'Item_Weight'] = value

#Show changes
get_missing_data_table(df)
```

```python
#----Outlet Size
df_copy2 = pd.concat([outlets_id, df['Outlet_Size']], axis='columns')
#Get null values
df_copy2 = df_copy2[df['Outlet_Size'].isnull()]
#Get how many nulls ids are
nulls = df_copy2['Outlet_Identifier'].unique()

from collections import Counter
for id in nulls:
    #Get value
    df_size = df.loc[outlets_id == id, 'Outlet_Size']
    value = Counter(df_size.values).most_common(1)[0][0]
    
    df.loc[(df_size == np.nan).index, 'Outlet_Size'] = value

#Show changes
get_missing_data_table(df)
```

With the new table we can see that the missing values in the weight of items were almost completely solved. However, there was no change in implementing the strategy with the size of the outlets. 

The four rows still remaining in 'Item_Weight' will be removed from the dataset.

```python
#Delete null in 'Item_Weight'
def get_null_observations(dataframe, column):
    return dataframe[pd.isnull(dataframe[column])]

def delete_null_observations(dataframe, column):
    fixed_df = dataframe.drop(get_null_observations(dataframe,column).index)
    return fixed_df

df = delete_null_observations(df, 'Item_Weight')

#Show results
get_missing_data_table(df)
```

In order to solve the missing values in 'Outlet_Size' I will then do an analysis to find a variable (without missing values) that explains the sales in each outlet quite well, with the purpose of using the average size in those variables. Some analyses of this type are those that were performed when the identifiers were explained within the DataFrame variables.

Let's look at the sales behavior in each outlet using the variable 'Outlet_Type'.

![Sales by Type](https://i.imgur.com/Q9rWsX6.jpg "Sales by Type")

As we can see, the quantity of products sold tends to be the same depending on the type of store. The strategy will then be to fill the missing values using the size that is most repeated according to the type of store.

![Outlet Size vs Outlet Type](https://i.imgur.com/L9IqC1K.jpg "Otlet Size by Outlet Type")

So the decision will be to fill 'Small' for grocery stores, 'Medium' for type 2 and 3 supermarkets, and 'Small' for type 1 supermarkets.

```python
#Get columns
size_type = pd.concat([df['Outlet_Size'], df['Outlet_Type']], axis='columns', keys=['Outlet_Size', 'Outlet_Type'])

#Transform
size_type.loc[size_type['Outlet_Type'] == 'Grocery Store', 'Outlet_Size'] = 'Small'
size_type.loc[size_type['Outlet_Type'] == 'Supermarket Type1', 'Outlet_Size'] = 'Small'
size_type.loc[size_type['Outlet_Type'] == 'Supermarket Type2', 'Outlet_Size'] = 'Medium'
size_type.loc[size_type['Outlet_Type'] == 'Supermarket Type3', 'Outlet_Size'] = 'Medium'

#Replace in DataFrame
df['Outlet_Size'] = size_type['Outlet_Size'].copy()

#Show results
get_missing_data_table(df)
```

## Exploratory Data Analysis

The idea behind the exploratory data analysis is to find patterns in the data. Since we want to build a model to predict sales of a product in a particular store, the patterns to observe are the ones that can explain this outcome in terms of some independent features.

The questions to solve with this EDA are:
* Buyers prefer heavier or ligher products?
* People prefer more healtier products?
* Are visual strategies efficient?
* People prefer price over quality?
* People prefer elder otlets?
* People prefer supermarkets or grocery stores?

### Buyers Prefer Heavier or Ligher Products?

A particular way to see if there is a pattern between the weight of the product and his sales in a particular store is a scatter plot. Plotting *Item_Weight* against *Item_Outlet_Sales* looks like the plot below.

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
sns.scatterplot(df['Item_Weight'], df['Item_Outlet_Sales'], ax=ax)
ax.set(title='Item Weight and Sales Correlation')
plt.show()
```

The scatter plot shows that it does not seem to exist a relation between the weight of a product and its success in sales. In the initial analysis phase the feature check showed outliers and, taking this scatterplot into account, it is very likely that this feature will be avoided in the regression model.

### People prefer more healtier products?

The initial analysis phase showed that the majority of the products in the dataset are low-fat products so the results for this question may have a bias. Basically, what we want is to find if there is being a low-fat product affects the number of sales of a specific product.

First, the *Item_Fat_Content* feature must be transformed using one hot encoding, a low-fat product will have a value of 0 and a regular fat one a value of 1.

```python
encode = {'Item_Fat_Content' : {'Low Fat': 0, 'Regular' : 1}}
df.replace(encode, inplace=True)
df.head(10)
```

Now, since *Item_Fat_Content* is a categorical feature and *Item_Outlet_Sales* is a continuous variable, a One-way ANOVA test will tell to us if there is a relation between variables.

```python
import scipy.stats as stats
low_fat = df[df['Item_Fat_Content'] == 0]['Item_Outlet_Sales']
regular_fat = df[df['Item_Fat_Content'] == 1]['Item_Outlet_Sales']

test_result = stats.f_oneway(low_fat, regular_fat)
print('F-statistic: {0}\np-value: {1}'.format(test_result[0],test_result[1]))
```

Since the p-value of the test is greater that threshold of 0.05, one can say that there were no statistically difference between group means. The differences in the number of sales means for low-fat products and regular products is not statistically significant.

The result suggests that people, although most of the products in the data set are low-fat, have no tendency to prefer such products.

### Are visual strategies efficient?

Exploring a possibly realtionship between the visibility of a product and its number of sales should tell us if buyers are affected by some kind of visibility trick, which is an important pattern to study for marketing.

Using a scatter plot the relationship between *Item_Visibility* and *Item_Outlet_Sales* is presented below.

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
sns.scatterplot(df['Item_Visibility'], df['Item_Outlet_Sales'], ax=ax)
ax.set(title='Visibility and Sales correlation')
plt.show()
```

Visibility does not show any linear correlation between the sales of tha product. This may means that the sample of buyers we have in the dataset are a type of buyers that is hard to persuade by just showing.

### People prefer price over quality?

To find price patterns and preferences the maximum retail price will be confront with sales. Basically, what a possible correlation could means is that buyers tend to buy products according to their price or popularity.

First, a scatterplot can give us an inital idea if there is any correlation.

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
sns.scatterplot(df['Item_MRP'], df['Item_Outlet_Sales'], ax=ax)
ax.set(title='Maximum Retail Price and Sales correlation')
plt.show()
```

The scatter plot shows a clear positive correlation. What this tells us is that the more MRP has a product (usually means more quality), the more sales (demand) this product has at the point of sale.

To explore this better, it is appropiate to make a scatter plot for each type of product.

```python
product_types = df['Item_Type'].unique()
fig = plt.figure(figsize=(15,10))
for product_index in range(len(product_types)):
    product = str(product_types[product_index])
    MRP = df[df['Item_Type'] == product]['Item_MRP']
    Sales = df[df['Item_Type'] == product]['Item_Outlet_Sales']
    ax = fig.add_subplot(4,4,product_index+1)
    sns.scatterplot(MRP, Sales, ax=ax)
    ax.set(title='{0} MRP and Sales correlation'.format(product))

plt.tight_layout()
plt.show()
```

The pattern seems to be the same for all type of products, maybe stronger in some products than others. To bring a correctly measure for the correlation, the Pearson's Correlation Coefficient is used and presented below.

```python
from scipy.stats import pearsonr
corr_df = pd.DataFrame()
corr_df['Item_Type'] = np.append(product_types,'TOTAL')
pearson_coef = []
pvalue = []
for product_index in range(len(product_types)):
    product = str(product_types[product_index])
    MRP = df[df['Item_Type'] == product]['Item_MRP']
    Sales = df[df['Item_Type'] == product]['Item_Outlet_Sales']
    pearson_coef.append(pearsonr(MRP, Sales)[0])
    pvalue.append(pearsonr(MRP, Sales)[1])
pearson_coef.append(pearsonr(df['Item_MRP'], df['Item_Outlet_Sales'])[0])
pvalue.append(pearsonr(df['Item_MRP'], df['Item_Outlet_Sales'])[1])
corr_df['Pearson_Coefficient'] = pearson_coef
corr_df['P-Value'] = pvalue
corr_df
```

The coefficients are from 0.5 to 0.6 in all product types, this is a moderate positive relationship between MRP and Sales. This may indicate a slighly preference to buy quality products rather than cheaper ones.

Again, a scatter plot between *Outlet_Establishment_Year* and *Item_Outlet_Sales* can give us an idea of a possible pattern.

### People prefer supermarkets or grocery stores?
Here, the pattern that we want to find is how likely are a product to be more selled, in a grocery store or a supermarket?. Since *Outlet_Type* is a categorical feature and *Item_Outlet_Sales* a continuous one, the One-way ANOVA test will tell us if the difference in sales is statistically significant.

```python
import scipy.stats as stats
grocery = df[df['Outlet_Type'] == 'Grocery Store']['Item_Outlet_Sales']
supermarket = df[df['Outlet_Type'] != 'Grocery Store']['Item_Outlet_Sales']

test_result = stats.f_oneway(grocery, supermarket)
print('F-statistic: {0}\np-value: {1}'.format(test_result[0],test_result[1]))
```

Grocery Stores and supermarket are different, but in the inital data analysis phase we saw 3 different types of supermarkets, the same ANOVA test will tell if there is any statistical difference between supermarkets.

```python
import scipy.stats as stats
supermarket1 = df[df['Outlet_Type'] == 'Supermarket Type1']['Item_Outlet_Sales']
supermarket2 = df[df['Outlet_Type'] == 'Supermarket Type2']['Item_Outlet_Sales']
supermarket3 = df[df['Outlet_Type'] == 'Supermarket Type3']['Item_Outlet_Sales']

test_result = stats.f_oneway(supermarket1, supermarket2, supermarket3)
print('F-statistic: {0}\np-value: {1}'.format(test_result[0],test_result[1]))
```

With a p-value less than critial threshold of 0.05, at least one type of supermarket is statistically different from the others. This test supports the hypotesis and plots of the initial data analysis phase where we found that type 3 supermarket sells are the highest.
